```python

#Korzystając z biblioteki numpy przygotuj dowolną macierz o wymiarach MxN.

#W Pythonie, korzystając z pętli (ver.1) i rozszerzonego operatora zakresu 
#(ver.2), przygotuj formuły dla:

#lustrzanego odbicia obrazu w poziomo

height = 9
width = 5
matrix = np.random.randint(0,10, size=(9,5))
print("oryginalna macierz\n",matrix)

# version 1

matrix_mirrored = np.empty([height,width], dtype = int)
for i in range(height):
    for j in range(width):
        matrix_mirrored[i][j] = matrix[height - i - 1 ][j]

print("macierz odwrocona w poziomo\n",matrix_mirrored)

# version 2

matrix_mirrored = np.empty([height,width], dtype = int)
matrix_mirrored = matrix[::-1]

print("macierz odwrocona wersja 2 poziomo\n",matrix_mirrored)
```

```python
#lustrzanego odbicia obrazu w pionowo

print("oryginalna macierz\n",matrix)

# version 1

matrix_mirrored = np.empty([height,width], dtype = int)
for i in range(height):
    for j in range(width):
        matrix_mirrored[i][j] = matrix[i][width - j - 1]

print("macierz odwrocona w pionie\n",matrix_mirrored)

# version 2

matrix_mirrored = np.empty([height,width], dtype = int)
matrix_mirrored = matrix[:,::-1]

print("macierz odwrocona wersja 2 pionowo\n",matrix_mirrored)


```

```python
#obrotu obrazu o 90° w prawo oraz 90° w lewo

print("oryginalna macierz\n",matrix)

# version 1 w prawo

matrix_rotate = np.empty([width, height], dtype = int)
for i in range(height):
    for j in range(width):
        matrix_rotate[j][i] = matrix[height - i - 1][j]

print("macierz odwrocona w prawo o 90\n",matrix_rotate)

# version 2
matrix_rotate = matrix[::-1]
matrix_rotate = matrix_rotate.T

print("macierz odwrocona wersja 2\n",matrix_rotate)


# version 1 w lewo

matrix_rotate = np.empty([width, height], dtype = int)
for i in range(height):
    for j in range(width):
        matrix_rotate[j][i] = matrix[i][width - j - 1]

print("macierz odwrocona w lewo 90\n",matrix_rotate)


# version 2 w lewo
matrix_rotate = matrix[:,::-1]
matrix_rotate = matrix_rotate.T

print("macierz odwrocona  w lewo wersja 2\n",matrix_rotate)



```

```python
#obrót obrazu o 180°

print("oryginalna macierz\n",matrix)

# version 1 w prawo

matrix_rotate = np.empty([height, width], dtype = int)
for i in range(height):
    for j in range(width):
        matrix_rotate[i][j] = matrix[height - i - 1][width - j - 1]

print("macierz odwrocona o 180 \n",matrix_rotate)

# version 2
matrix_rotate = np.empty([height, width], dtype = int)
matrix_rotate = matrix[::-1,::-1]
print("macierz odwrocona o 180 wersja 2\n",matrix_rotate)


```

```python
#rozszerzenia macierzy do kwadratu MxM (jeżeli M>N) - kolumny po lewej i prawej stronie są wyzerowane
print("macierz glowna\n",matrix)
matrix_square = np.zeros((height,height), dtype=int)
difference = height - width 
difference = difference // 2
for i in range(height):
    for j in range(width):
            matrix_square[i][j + difference] = matrix[i][j]

print("macierz wynikowa \n",matrix_square)

matrix_square = np.zeros((height,height), dtype=int)
matrix_square[:, difference:difference + width] = matrix
print("version 2\n",matrix_square)

```

```python

#wycięcia z macierzy kwadratu NxN (jeżeli M>N) - wycinamy środkową część macierzy
#Wyniki zaprezentuj na wybranej macierzy numpy o wymiarach 9x5 printując jej zawartość w arkuszu Jupyter


difference = height - width
difference = difference // 2
matrix_square = np.empty((width,width), dtype=int)
if( height > width):
    for i in range(width):
        for j in range(width):
            matrix_square[i][j] = matrix[i+difference][j]

    print("macierz oryginalna\n",matrix)
    print("macierz wycieta\n",matrix_square)


matrix_square = np.empty((width,width), dtype=int)
matrix_square[:][:] = matrix[2:7][:]

print("macierz wycieta v2\n",matrix_square)
```

```python

img = (cv2.imread("pies.jpeg") / 255.0).astype(np.float32)

crop_height = 800
crop_width = 1000
new_size = (crop_width, crop_height) 

cropped_img = cv2.resize(img, new_size)
cropped_img = cv2.cvtColor(cropped_img, cv2.COLOR_BGR2RGB)


x_center = crop_width // 2
y_center = crop_height // 2

def mask_create(triangle):
    mask = np.zeros((crop_height, crop_width,3), dtype=np.float32)
    cv2.fillPoly(mask, [triangle],(1,1,1))
    return mask

mask_table = []
pts = np.array([[0, 0], [x_center, y_center], [crop_width, 0]], np.int32)
mask_table.append(mask_create(pts))

pts = np.array([[x_center, y_center], [crop_width, crop_height], [0, crop_height]], np.int32)
mask_table.append(mask_create(pts))

pts = np.array([[crop_width, 0], [x_center, y_center], [crop_width, crop_height]], np.int32)
mask_table.append(mask_create(pts))

pts = np.array([[0, 0], [x_center, y_center], [0, crop_height]], np.int32)
mask_table.append(mask_create(pts))


# rand_table = [ i for i in range (4)]

# if random.choice([True, False]):
#     rand_table[0], rand_table[1] = rand_table[1], rand_table[0]

# if random.choice([True, False]):
#     rand_table[2], rand_table[3] = rand_table[3], rand_table[2]

# print(rand_table)


image_shuffled = np.zeros((crop_height, crop_width,3), dtype= np.float32)



number = np.random.randint(0,2)
if number == 1:
    masked_image = cropped_img * mask_table[0]
    image_shuffled += cv2.flip(masked_image,0) 
    masked_image = cropped_img * mask_table[1]
    image_shuffled += cv2.flip(masked_image,0)
else:
    masked_image = cropped_img * mask_table[0]
    image_shuffled += masked_image 
    masked_image = cropped_img * mask_table[1]
    image_shuffled += masked_image  


number = np.random.randint(2,4)
if number == 2:
    masked_image = cropped_img * mask_table[2]
    image_shuffled += cv2.flip(masked_image,1)
    masked_image = cropped_img * mask_table[3]
    image_shuffled += cv2.flip(masked_image,1)
else:
    masked_image = cropped_img * mask_table[2]
    image_shuffled += masked_image
    masked_image = cropped_img * mask_table[3]
    image_shuffled += masked_image  
    

image_shuffled = np.clip(image_shuffled, 0.0, 1.0)
plt.imshow(image_shuffled)
plt.show()



```
